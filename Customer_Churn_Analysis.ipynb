```python
df["TotalCharges"]=pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Churn",data=df)
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.savefig("Graphs/churn_distribution.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="gender",hue="Churn",data=df)
plt.title("Gender vs Churn")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.savefig("Graphs/gender_vs_churn.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Contract",hue="Churn",data=df)
plt.title("Contract Type vs Churn")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.savefig("Graphs/contract_vs_churn.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="InternetService",hue="Churn",data=df)
plt.title("Internet Service vs Churn")
plt.xlabel("Internet Services")
plt.ylabel("Number of Customers")
plt.savefig("Graphs/internet_service_vs_churn.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["MonthlyCharges"],bins=30,kde=True)
plt.title("Distribution of Monthly Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.savefig("Graphs/monthly_charges_distribution.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["tenure"],bins=30,kde=True)
plt.title("Customer Tenure Distribution")
plt.xlabel("Tenure (Months)")
plt.ylabel("Number of Customers")
plt.savefig("Graphs/tenure_distribution.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x="PaymentMethod",hue="Churn",data=df)
plt.title("Payment Method vs Churn")
plt.xlabel("Payment Method")
plt.ylabel("Number of Customers")
plt.xticks(rotation=20)
plt.savefig("Graphs/payment_method_vs_churn.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="SeniorCitizen",hue="Churn",data=df)
plt.title("Senior Citizen vs Churn")
plt.xlabel("Senior Citizen (0 No, 1 = Yes)")
plt.ylabel("Number of Customers")
plt.savefig("Graphs/senior_citizen_vs_churn.png",dpi=300,bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
numeric_df=df.select_dtypes(include=['int64','float64'])
sns.heatmap(numeric_df.corr(),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("Graphs/correlation_heatmap.png",dpi=300,bbox_inches="tight")
plt.show()
```

# Final Key Insights

1. Most customers stayed with the company, while a smaller percentage churned.

2. Customers with month-to-month contracts showed the highest churn.

3. Fiber optic internet users had higher churn compared to DSL users.

4. Customers using Electronic Check showed higher churn.

5. Customers with longer tenure were generally more loyal.

6. Gender did not have a significant impact on customer churn.

7. Senior citizens showed different churn patterns compared to non-senior citizens.

8. Data visualization helped identify important business trends and customer behavior.
